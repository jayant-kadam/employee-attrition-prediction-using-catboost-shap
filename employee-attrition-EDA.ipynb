```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/train.csv")
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

### Attrition Distribution

```python
df["Attrition"].value_counts().plot(
    kind="bar",
    title="Employee Attrition Distribution"
)

plt.show()
```

### Numerical Features Distribution

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols].hist(
    figsize=(15,10),
    bins=20
)

plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df[num_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

### Attrition vs Overtime

```python
sns.countplot(
    x="Overtime",
    hue="Attrition",
    data=df
)

plt.show()
```

### Attrition vs Job Satisfaction

```python
sns.countplot(
    x="Job Satisfaction",
    hue="Attrition",
    data=df
)

plt.xticks(rotation=45)
plt.show()
```

### Attrition vs Work-Life Balance

```python
sns.countplot(
    x="Work-Life Balance",
    hue="Attrition",
    data=df
)

plt.show()
```

### Attrition vs Job Role

```python
plt.figure(figsize=(12,6))

sns.countplot(
    y="Job Role",
    hue="Attrition",
    data=df
)

plt.show()
```
